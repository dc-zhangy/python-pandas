# 缺失值处理


```python
import numpy as np
df = pd.DataFrame([[1., 6.5, 3.], [2., np.nan, np.nan],
                     [None, np.nan, np.nan], [np.nan, 9, 5.]])
df
```

## 查看缺失情况

```python
df.isnull().sum(axis=0)
```

## 删除


```python
#删除有缺失值的行
df.dropna()
```

```python
#删除所有值为缺失值的行/列
df.dropna(how='all',axis=0)
```

```python
#删除缺失值个数大于等于阈值的的行、列
df.dropna(thresh=2)
```

```python
#删除某些列有缺失值的行
df.dropna(subset=[1])
```

## 填充

```python
import numpy as np
df = pd.DataFrame([[1., 6.5, 3.], [2., np.nan, np.nan],
                     [None, np.nan, np.nan], [np.nan, 9, 5.]])
df
```

```python
#常数填充
df.fillna(10)
```

```python
#周围值填充
df.fillna(method='ffill',axis=0)
```

```python
#统计特征填充
df.fillna(df.median())
```

## 利用函数或映射进行数据转换

```python
df = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon', 'Pastrami', 'corned beef', 'Bacon',
                              'pastrami', 'honey ham','nova lox'],
                    'ounces': [4, 3, 12, 6, 7.5, 8, 3,5, 6]})
df
```

```python
#map
meat_to_animal = {
'bacon': 'pig',
'pulled pork': 'pig',
'pastrami': 'cow',
'corned beef': 'cow',
'honey ham': 'pig',
'nova lox': 'salmon'
}

lowercased = df['food'].str.lower()
lowercased.map(meat_to_animal)
```

```python
#replace
df['food'].str.lower().replace(meat_to_animal)
```
